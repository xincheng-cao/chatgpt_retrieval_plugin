```python
def query_all(collection:Collection):
    output_fields=[e.name for e in collection.schema.fields]

    res=collection.query(
        expr='url in ["",]',
        output_fields=output_fields,
    )
    def get_key_fn(e):
        return e['id']
    res.sort(reverse=False,key=get_key_fn)
    return res
```

```python
query_all(col_ctrip_geo)
```
